```python
df=df.drop(['id_x','car_rel_url_x','datetime_scrape','name','price_x','currency_x','datetime_product',
            'city','day','hour','attributes','engine_displacement_unit','kilometrage_unit','barter',
            'loan','salon','spare_parts','vip','featured','img_url','id_y','cars_id','cars_id','car_rel_url_y',
            'datetime','description','owner_name','shop_name','phone','updated','views','vin','Buraxılış ili',
            'Marka','Model','Sahiblər','Vəziyyəti','Şəhər','Şəhər','car_details_id_y','car_rel_url','car_details_id_x'],axis=1)
```

```python
#valuyta USD olanlar manata cevrilir
df['price_AZN'] = np.where(df['currency_y'] == 'USD', df['price_y'] * 1.7,
                  np.where(df['currency_y'] == 'EUR', df['price_y'] * 1.85, df['price_y']))
```

```python
#yanacag novlerin cixarir, yeni sutuna elave dilir
df['yanacaq_novu']=df['Mühərrik'].str.extract('(Dizel|Benzin|Hibrid|Elektrik)')
```

```python
#at gucunu elave sutun kimi cixarir
df['at_gucu']=df['Mühərrik'].str.extract(r'(\d+)\s*a\.g\.')
#a.g olan hisseni silir
df['at_gucu']=df['at_gucu'].str.replace(r'\s*/\s*\d+\s*a\.g\.', '', regex=True)
df['at_gucu'] = df['at_gucu'].astype(float)
```

```python
df.isna().sum()
```

```python
# Yürüş sutununda km silir
df['yürüş']=df['Yürüş'].str.replace('km','').str.replace(' ','').astype(int)
```

```python
# Ban novu elava sutun olacaq cixarilir
df['ban_novu']=df['Ban növü'].str.extract('(Hetçbek|Offroader|Sedan|Liftbek|Yük maşını|Kompakt-Van|Furqon|Minivan|Mikroavtobus|Universal|Motosiklet|Kupe|Pikap)')
```

```python
df = df.drop_duplicates()
```

```python
df=df.drop(['currency_y','Mühərrik','Hansı bazar üçün yığılıb','Yürüş','extra_info',
            'Ban növü','Yerlərin sayı','price_y','Qəzalı','Rəng'],axis=1)
```

```python
df = df.drop_duplicates()
df
```

```python
X=df.drop('price_AZN',axis=1)
y=df['price_AZN'].copy()
```

# LinearRegression

```python
#train ve teste bolunur
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,random_state=42)
```

```python
# cat ve num features yaradilir
cat_features=X_train.select_dtypes(exclude=[np.number]).columns
num_features=X_train.select_dtypes(include=[np.number]).columns
```

```python
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.ensemble import BaggingRegressor
from sklearn.model_selection import GridSearchCV
```

```python
cat_pipeline=Pipeline([
    ('imput',SimpleImputer(strategy='constant', fill_value='Missing')),
    ('onehot', OneHotEncoder(handle_unknown ='ignore',sparse_output=False))
])
num_pipeline=Pipeline([
    ('imputer',SimpleImputer(strategy='median')),
    ('scale',StandardScaler())
])

transformer = ColumnTransformer ([
    ('num',num_pipeline,num_features),
    ('cat',cat_pipeline,cat_features)
], remainder='passthrough')


estimator=LinearRegression()

full_pipeline=Pipeline([
    ('preprocessing',transformer),
    ('estimator',estimator)
])
full_pipeline.fit( X_train,y_train)
```

```python
full_pipeline.score(X_train,y_train)
```

```python
full_pipeline.score(X_test,y_test)
```

# Bagging

```python
cat_pipeline=Pipeline([
    ('imput',SimpleImputer(strategy='constant', fill_value='Missing')),
    ('onehot', OneHotEncoder(handle_unknown ='ignore',sparse_output=False))
])
num_pipeline=Pipeline([
    ('imputer',SimpleImputer(strategy='median')),
    ('scale',StandardScaler())
])

transformer = ColumnTransformer ([
    ('num',num_pipeline,num_features),
    ('cat',cat_pipeline,cat_features)
], remainder='passthrough')


estimator=BaggingRegressor(n_estimators=100,
                           bootstrap=True)

full_pipeline=Pipeline([
    ('preprocessing',transformer),
    ('estimator',estimator)
])
full_pipeline.fit( X_train,y_train)
```

```python
full_pipeline.score(X_train,y_train)
```

```python
import tensorflow as tf
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.layers import Dense
```

```python
cat_pipeline=Pipeline([
    ('imput',SimpleImputer(strategy='constant', fill_value='Missing')),
    ('onehot', OneHotEncoder(handle_unknown ='ignore',sparse_output=False))
])
num_pipeline=Pipeline([
    ('imputer',SimpleImputer(strategy='median')),
    ('scale',StandardScaler())
])

transformer = ColumnTransformer ([
    ('num',num_pipeline,num_features),
    ('cat',cat_pipeline,cat_features)
], remainder='passthrough')

X_train_transformed=transformer.fit_transform(X_train)
X_test_transformed=transformer.transform(X_test)

model= tf.keras.Sequential([
    tf.keras.layers.Input(shape=(X_train_transformed.shape[1:])),
    Dense(64,activation='relu'),
    Dense(32,activation='relu'),
    Dense(10,activation='relu'),
    Dense(1,activation='linear')
])


model.compile(loss='mse', optimizer=Adam(learning_rate=0.001),metrics=['mae'])
history=model.fit(X_train_transformed,y_train,epochs=50)
```

```python
from sklearn.ensemble import VotingRegressor
from sklearn.ensemble import GradientBoostingRegressor


cat_pipeline=Pipeline([
    ('impute',SimpleImputer(strategy='constant',fill_value='Missing')),
    ('onehot',OneHotEncoder(handle_unknown='ignore',sparse_output=False))
])

num_pipeline=Pipeline([
    ('impute',SimpleImputer(strategy='median')),
    ('scale',StandardScaler())
])

transformer=ColumnTransformer([
    ('num',num_pipeline,num_features),
    ('cat',cat_pipeline,cat_features)
],remainder='passthrough')

voting=VotingRegressor(
    estimators=[
        ('lnr',LinearRegression()),
        ('rfr',RandomForestRegressor(random_state=42)),
        ('gbr',GradientBoostingRegressor(random_state=42))
        ]
)

full_pipeline=Pipeline([
    ('preprocessing',transformer),
    ('voting',voting)
])


full_pipeline.fit(X_train[:5000],y_train[:5000])
```
